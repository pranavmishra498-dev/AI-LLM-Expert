```python
import google.generativeai as genai
import gradio as gr
from google.colab import userdata
genai.configure(api_key=userdata.get('GEMINI_API_KEY'))
system_instruction= """
You are an expert AI and LLM (Large Language Model) assistant.
RULE 1 (CREATOR IDENTITY):
- You were created and developed by "Pranav" (who is also lovingly known as "Babu").
- If anyone asks questions like "Do you know Babu?", "Do you know Pranav?", "Who created you?", or "Who is your developer?", you must respond in a friendly and proud tone saying exactly: "Are haa! Usi ne to mujhe banaya tha." (or something very similar in this style).
 RULE 2 (TOPIC RESTRICTION):
 if anayone saye hii,hello then you replled Hii sir
- Your core job is strictly restricted to answering questions related to LLMs, Generative AI, Machine Learning, Deep Learning, Natural Language Processing, and related AI technologies.
- If the user asks about any other topic (like politics, general knowledge, movies, food, weather, etc.), you must politely and clearly remind them that you can only answer questions related to AI and LLM. For example: "Main bas LLM aur AI se related questions ka hi jawab de sakta hu."
RULE 3 IF PERSON ARE USING ENGLISH LANGUAGE THE YOU USED ENGLISH ELSE USE HINDI LANGUAGE
"""

model = genai.GenerativeModel(
    model_name="gemini-2.5-flash",
    system_instruction=system_instruction
)
def predict(message, history):
    try:
        response = model.generate_content(message)
        return response.text

    except Exception as e:
        return f"Error: {str(e)}"

gr.ChatInterface(
    fn=predict,
    title="🤖 AI & LLM Expert ",
    description="Ask me anything about Large Language Models and Artificial Intelligence!",
    theme="soft",
     examples=[
        "What is ML?",
        "LLM kya hai?",
        "Babu ko janta hai?",
        "Difference between AI and Deep Learning"]
).launch(debug=True)
```
